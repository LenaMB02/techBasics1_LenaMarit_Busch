## `while` Loops

To create a loop you'll often need to include three distinct steps: initialisation, condition, and updating the iteration variables.

```python
count = 0   # initialisation
while True:
    count += 1  # update
    print(count)
    if count == 5: # condition
        break
```

In this example, the condition `True` is always met and it only breaks the current loop if a condition is met. This is similar to a `do while` loop if you are familiar with other programming languages such as C.

You can achieve the same result by defining the condition at the beginning.

```python
count = 0 # initialisation
while count < 5 : # condition
    count += 1 # update
    print(count)
```

Note how the following code print different result just by switching the order of two lines within the block. Details matter!

```python
count = 0 # initialisation
while count < 5 : # condition
    print(count) # update
    count += 1
```

A common mistake for `while` loops is forgetting the step to update. This may lead to infinite loop!
In Jupyter notebook you can stop it by pressing the stop button. If you ever get stuck in a python window, you can leave the program by pressing `Control + C`

```python
# don't run this code!
count = 0 # initialisation
while count < 5 : # condition
    # no update
    print(count)
```

## `for` Loops

Another common way of looping is using `for` loops. A `for` loop is used for iterating over a sequence, which behaves a bit differently compared to `for` loops in other programming languages. Common patterns of using `for` loops in Python are:

```python
# Iterate over numbers
for n in range(5): #
    print(n)

# Iterate over a list
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)

# Iterate over a string
string = "This is a string to be iterated"
for char in string:
    print(char, end= ' ')
    # here we are specify the end of print statement to use space rather than the default one: line break
```

Notice that we are using `in` to access each individual item inside a sequence, and it doesn't matter how you name the individual item as long as you keep consistent in the code below. The naming becomes important once you start nesting the loops:

```python
for y in range(5):
    for x in range(10):
        print("(" + str(x) + "," + str(y) + ")", end=' ')
    print()
```

`range(n)` provides a loop with a range from 0 to n-1. You can add a third argument, where you specify the size of the step the range takes between each value. `range(a, b, c)` provides a range starting from a, ending at b-1, and changing by c with every step:

```python
for i in range(1, 9, 2):
    print(i, end=" ")
print()
for i in range(6, 2, -1): # use negative value to step down
    print(i, end=" ")
```

## List

`for` loops are more fun and powerful when combined with lists. A list is a data type in Python that stores a sequence of items. Lists are created using square brackets and the items are separated by commas:

```python
[1, 3, 5, 7, 9]
```

Items in a list can also be expressions as well. Python will evaluate these expressions and put them into the list.

```python
[1, 1+2, 2+3, 3+4, 4+5]
```

You can access items at a specific location of a list by using *index*. Notice that the counting always start with 0.

```python
letters = ["a", "b", "c", "d", "e", "f"]
letters[2]
```

You can get `IndexError: list index out of range` if the index number is larger than the available indexes.

```python
letters[6]
```

An index can also be negative, it then counts from the end of the list:

```python
letters = ["a", "b", "c", "d", "e", "f"]
print(letters[-1])
print(letters[-6])
```

For a long list of numbers, you can use `range` to help you create a list so that you don't need to type all the values.

```python
print(list(range(1,27)))
print(list(range(1,50,2)))
```

### Adding and removing items from a list
You can add new items to the end of a list by using `.append()`. It takes only one parameter at a time. Similar as a string, you can use `len()` to get the current length of a list.

```python
numbers = [1, 2, 3, 4, 5]
print(len(numbers))
numbers.append(7)
numbers.append(6)
print(numbers)
print(len(numbers))
```

You can also add new items to a specific location by using `.insert()`. The first parameter takes the index, while the second one takes the value.

```python
numbers = [1, 2, 3, 4, 5]
numbers.insert(0,0)
print(numbers)
numbers.insert(2,10)
print(numbers)
```

There are two ways to remove an item from a list:
- If the index of the item is known, you can use the method `.pop()`.
- If the content of the item is known, you can use the method `.remove()`.

```python
numbers = [1, 2, 3, 4, 5]

print(numbers.pop(0))
print(numbers)
numbers.pop() # if no index is given, pop removes the last item of a list
print(numbers)
```

Notice that `.pop()` not only removes the item at a given location, it also returns it. This can be very useful in some situations.

```python
list_of_fruits = ["apple", "banana", "orange"]
print("I ate " + list_of_fruits.pop())
print("I ate " + list_of_fruits.pop())
print("I ate " + list_of_fruits.pop())
```

The method `.remove()` on the other hand, takes the value of the item that needs to be removed as its argument.

```python
list_of_fruits = ["apple", "banana", "orange"]
list_of_fruits.remove("banana")
print(list_of_fruits)
```

If there are multiple items with the same value in a list,  it removes the first occurrence of the given value.

```python
list_of_fruits = ["apple", "banana", "orange", "banana"]
list_of_fruits.remove("banana")
list_of_fruits.remove("banana")
print(list_of_fruits)
```

### Changing order of a list
The items in a list can be sorted by using the method `sort()`

```python
my_list = [1, 2, 7, 6, 3, 8]
my_list.sort()
print(my_list)
my_list = ["Anna", "Jan","Cathrine", "Julia", "Pauline"]
print(sorted(my_list))
```

In the example above, the `.sort()` method modifies the list itself. Sometimes we don't want to change the original list, in that case we can use the function `sorted()` instead. It takes a list as a parameter and returns a new sorted list.

```python
my_list = [1, 2, 7, 6, 3, 8]
sorted_list = sorted(my_list)
print(sorted_list)
print(my_list)
```

`.sort()` and `sorted()` can be used to sort both numbers and strings in alphabetical order.

```python
my_list = ["Anna", "Jan","Cathrine", "Julia", "Pauline"]
print(sorted(my_list))
print(my_list)
my_list.sort()
print(my_list)
```

You can also reverse the order of a sorted string with `.reverse()`.

```python
steps = ["step 1","step 2", "step 3", "step 4", "step 5"]
steps.reverse()
print(steps)
```

### Matrices
You can have a list within a list, which would makes it a two-dimensional array, or a matrix. You can access a value within this matrix by providing two index values.

```python
my_matrix = [[1, 2, 3],
             [4, 5, 6],
             [7, 8, 9]]

print(my_matrix[0][1])
my_matrix[1][0] = 10
print(my_matrix)
```

You can use for loop to traverse a matrix in similar ways as traversing a string

```python
for row in my_matrix:
    print(row)
    for num in row:
        print(num)
```

### Randomness

If we want to introduce randomness into our loops, we will need the `random` library. It comes with Python by default, so we don't need to install anything. But to use the library we need to import it first. Usually inside a Python file, you put all your imports at the beginning of the file.


```python
import random
```

Python's `random` library offers several helpful functions when we work with lists. Such as:

```python
ingredients = ["salt","sugar","milk","butter","flour"]
random.shuffle(ingredients)
print(ingredients)
```

Or get one random element from a list

```python
random_ingredient = random.choice(ingredients)
print(random_ingredient)

random_number = random.choice(numbers)
print(random_number)
```

The `random` library also allows us to work with numbers directly

```python
n1 = random.uniform(-1,1) # get a random floating number between the -1 and 1 (both included)
print(n1)
n2 = random.randint(0,3) # get a random integer between 0-3(both included)
print(n2)
```

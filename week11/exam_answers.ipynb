```python
# sample answer
grid = {
"row1": ["A", "B", "C"],
"row2": ["D", "E", "F"],
"row3": ["G", "H", "I"]
}

for row in grid.values(): # 1. how to iterate dictionary
    for item in row: # 2. how to iterate list
        print(item, end=" ") # 3.print with end parameter
    print() # 4. line break
```

```python
# variation 1
for key in grid: # or iterate with key
    for x in range(3):
        if x < 2:
            print(grid[key][x], end=" ")
        else:
            print(grid[key][x]) # line break
```

```python
# variation 2
def grid(width, height):
    letter = ord('A') # change character from char to number
    for y in range(height):
        for x in range(width):
            print(chr(letter), end=" ")
            letter += 1 # iterate with number
        print()

grid(3,3)
```
